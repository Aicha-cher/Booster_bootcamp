## Q. Write a function that converts slowness (us/m) into velocity (m/s)

```python
def slow2vel(curve):
    if curve.units.upper == 'US/M':
        converted = 1e6/curve 
    if curve.units.upper == 'US/F' or 'US/FT':
        scalar = 3.2808  # number of feet per metre
        converted = 1e6/(curve*scalar)
    return converted
```

```python
# Create a Vp/Vs vs Acoustic Impedance plot
Vp = slow2vel(Pslow)
AI = Vp * RHOB
```
